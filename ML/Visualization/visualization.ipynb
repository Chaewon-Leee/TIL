# Pandas matplotlib
- 데이터간의 상관관계를 볼 때 scatter graph 사용 가능

```python
import numpy as np
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
results = pd.read_csv("drive/MyDrive/ML/Visualization/viz/housing.data", header=None) # sep : seperate는 빈공간으로 지정
result = results[0].str.split()
```

```python
columnss = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
example = pd.Series(data=np.array(result)[0], index=columnss, dtype=float)
df = pd.DataFrame(example).T

for i in range(1,len(result)):
  temp = pd.Series(data=np.array(result)[i], index=columnss, dtype=float)
  df = pd.concat([df, pd.DataFrame(temp).T])
df.reset_index(inplace=True, drop=True)
```

```python
df
```

```python
fig = plt.figure()
ax = []
df_data = df

for i in range(1, 5):
  ax.append(fig.add_subplot(2,2,i))

ax[0].scatter(df_data['CRIM'], df_data['MEDV'], s=3)
ax[1].scatter(df_data['PTRATIO'], df_data['MEDV'], s=3)
ax[2].scatter(df_data['AGE'], df_data['MEDV'], s=3)
ax[3].scatter(df_data['NOX'], df_data['MEDV'], s=3)

for i in range(4):
  ax[i-1].get_xaxis().set_visible(False)
  ax[i-1].get_yaxis().set_visible(False)

plt.show()
```

```python
fig = plt.figure()
ax = []
df_data = df

for i in range(1, 5):
  ax.append(fig.add_subplot(2,2,i))

columns = ['CRIM', 'PTRATIO', 'AGE', 'NOX']
colors = ["b", "g", "c", "r"]

for i in range(4):
  ax[i].scatter(df_data[columns[i]], df_data['MEDV'],
                s=3, color=colors[i], label=columns[i])
  ax[i].legend()
  ax[i].set_title(columns[i])
  ax[i].get_yaxis().set_visible(False)
  ax[i].get_xaxis().set_visible(False)

plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

```python
sns.set(style="white", palette="muted", color_codes=True)
rs = np.random.RandomState(10)

f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
sns.despine(left=True)

d = rs.normal(size=100)

sns.distplot(d, kde=False, color="b", ax=axes[0, 0])
sns.distplot(d, hist=False, rug=True, color="r", ax=axes[0, 1])
sns.distplot(d, hist=False, color="g", kde_kws={'shade':True}, ax=axes[1, 0])
sns.distplot(d, color="m", ax=axes[1, 1])

plt.setp(axes, yticks=[])
plt.tight_layout()
```
