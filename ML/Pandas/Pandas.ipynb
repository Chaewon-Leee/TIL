## lambda, map apply

lambda
- 한 줄로 함수를 표현하는 익명 함수 기법
- lisp 언어에서 시작됨
- lambda 변수: 변수 식
  - 삼항연산자와 결합하여 사용 가능

map 함수
- 함수와 sequence형 데이터를 인자로 받아 각 element마다 입력받은 함수를 적용하여 list로 반환
- 일반적으로 함수를 lambda 형태로 표현!
- map(함수, sequence)

map for series
- Pandas의 series type의 데이터에도 map 함수 사용 가능
- function 대신 dict, sequence형 자료 등으로 대체 가능
  - series.map(함수) : 해당 series에 함수 적용
  - series.map(dict / series) : 해당 sequence와 동일한 인덱스에 값 바꿔치기
    - 나머지 값을 NaN 처리

replace function
- map 함수의 기능 중 데이터 변환 기능만 담당
- 데이터 변환시 많이 사용
- replace(dict/list1, list2) : map과 동일하게 dict 타입으로 부여 혹은 타겟 리스트와 변경할 리스트를 추가하여 변환 가능


apply & applymap
- map과 달리, series 전체 (col)에 해당 함수를 적용
- 입력값이 series 데이터로 입력받아 handling 가능
  - df.apply(함수) : 각 열별로 적용한 함수 값 추출
    - 시리즈 데이터로 출력함!
  - .applymap(f) : 시리즈 단위가 아닌 요소 단위로 적용

```python
ex = [1,2,3,4,5]
f=lambda x:x+2
list(map(f, ex)) # lambda 함수를 입력받아 ex 리스트에 적용
# 리스트를 안할 경우, 메모리 주소만 나오게 됨
```

```python
list(map(lambda x : x+x, ex)) # 익명함수 바로 사용 가능
```

## Groupby
- SQL groupby 명령어와 동일
- split -> apply -> combine 과정을 거쳐 연산
- df.groupby("묶음의 기준이 되는 컬럼")["적용받는 컬럼"].연산()
  - df.groupby("Team")["Points"].sum() : Team을 기준으로 묶고, Points를 sum해서 보여줘라 => Series 데이터로 반환
- 한 대 이상의 col을 묶을 수 있다

Hierarchical index
- Groupby 명령의 결과물도 결국은 데이터프레임!
- 두 개의 col으로 groupby를 할 경우, 인덱스가 두 개 생성됨
- unstack() : 그룹으로 묶여진 데이터를 매트릭스 형태로 전환해준다
- swaplevel() : 인덱스의 순서 서로 변환
  - sortlevel(0)과 함께 사용시 정렬 적용됨!
- operations : Index level을 기준으로 연산 수행 가능
  - .sum(level=0)

Groupby2
- split된 상태 즉, Grouped 상태를 추출 가능
  - .get_group() : 특정 그룹의 상태 확인 가능
- Aggregation
  - 추출된 group 정보에는 세 가지 유형의 apply 가능
    1. Aggregation : 요약된 통계정보 추출
    2. Transformation : 해당 정보 변환
    3. Filtration : 특정 정보를 제거하여 보여주는 필터링 기능
- Transformation
  - Aggregation과 달리 key값 별로 요약된 정보가 아님
  - 개별 데이터의 변환을 지원함 -> 단 max나 min처럼 series 데이터에 적용되는 데이터들은 key 값을 기준으로 grouped된 데이터 기준
  - .transform(함수)
- filter
  - 특정 조건으로 데이터를 검색할 때 사용
  - .filter(함수식) : filter안에 bool조건 존재 필요

Transformation

```python
df
```

```python
score = lambda s:(s)
grouped.transform(score) # 개별 데이터에 대해 함수가 적용돼서 나온 것 볼 수 있다
```

```python
score = lambda s:(s.max())
grouped.transform(score)
# max의 경우 시리즈 데이터에 적용되기 때문에 개별 데이터가 아닌 key값 기준인 grouped된 데이터에 적용된다
```

```python
score = lambda x: (x - x.mean()) / x.std()
grouped.transform(score) # 그룹별로 정규화
```

## Pivot table & Crosstab
- Pivot table
  - 엑셀과 역할이 동일
  - index축은 groupby와 동일하게 적용 (묶어주는 개념), col에 추가로 labelling 값을 추가
  - value에 numeric type 값을 aggregation하는 형태
- crosstab
  - 특히 두 컬럼에 교차 빈도, 비율, 덧셈 등을 구할 때 사용
  - Pivot table의 특수한 형태
  - User-Item Rating Matrix 등을 만들 때 사용 가능

pivot table

```python
import pandas as pd
import dateutil
```

```python
df_phone = pd.read_csv("phone_data.csv")
df_phone['date'] = df_phone['date'].apply(dateutil.parser.parse, dayfirst=True)
df_phone.head()
```

```python
df_phone.pivot_table(['duration'], # duration을 가로축 = duration을 그룹화하겠다
                     index=(df_phone.month, df_phone.item), # month, item을 세로축
                     columns=df_phone.network, aggfunc="sum", fill_value=0) # network을 기준으로 데이터를 나눈다
                     # aggregation은 sum으로 계산, NaN값은 0으로 채움
```

crosstab

```python
pd.crosstab(index=df_phone.month, columns=df_phone.item, values=df_phone.duration, aggfunc="sum").fillna(0)
# item 가로축으로, month는 세로축으로 해서 duration 값으로 crosstab 확인
```

```python
df_phone.groupby(['month', 'item'])['duration'].sum().unstack()
```

```python
df_phone.pivot_table(["duration"], index=df_phone.month, columns=df_phone.item,
                     aggfunc="sum", fill_value=0) # pivot table로도 가능!
```

## Merge & Concat

- Merge : 두 개의 데이터를 하나로 합침
- concat : 같은 형태의 데이터를 붙이는 연산작업 = append라고 생각하면 된다

Merge

```python
import pandas as pd
```

```python
raw_data = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_score': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
df_a = pd.DataFrame(raw_data, columns = ['subject_id', 'test_score'])
df_a
```

```python
raw_data = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df_b = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_b
```

```python
pd.merge(df_a, df_b, on="subject_id") # subject id를 기준으로 동일하게 존재하는 데이터에 대해 값 합친다
```

```python
pd.merge(df_a, df_b, left_on='subject_id', right_on='subject_id') # 데이터프레임이 컬럼 이름이 다를 경우, 헤딩 두 개의 열의 값을 비교하여 병합하게 됨
```

```python
pd.merge(df_a, df_b, on='subject_id', how='left') # how로 지정 가능 
```

```python
pd.merge(df_a, df_b, on='subject_id', how='right')
```

```python
pd.merge(df_a, df_b, on='subject_id', how='outer')
```

```python
pd.merge(df_a, df_b, on='subject_id', how='inner')
```

```python
pd.merge(df_a, df_b, right_index=True, left_index=True) # 인덱스를 기준으로 조인
```
