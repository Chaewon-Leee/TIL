## Why and What deep learning

딥러닝
- 이전에는 데이터에서 사람이 직접 룰을 찾아서 업데이트하는 형식으로 발전해왔다
    - 진돗개는 하얀색이다 + 노란색이다 + .... 업데이트
- 사람이 모든 규칙을 찾긴 어렵다!
- 컴퓨터가 이런 규칙을 찾는 것이다 = 머신러닝
    - 이전에는 데이터와 룰을 가지고 답을 찾는 거였다면, 현재는 답과 데이터를 줄 경우, 머신이 룰을 찾는 형식으로 바뀌게 된다
        - 진돗개다. 개의 색상, 눈동자, 코, 수염 등...
        - 이렇듯 머신에세 학습을 시킬 때, 정답과 feature을 정해서 알려준다
        - 그러나 사람이 못보는 것, 모르는 것에도 패턴이 있다
        - 머신이 전부 다 보되 그 중에서 스스로 중요한 것만 강조하면 안될까 = 딥러닝
- 전통적인 머신러닝은 인풋과 중요한 특성들을 미리 알려줬는데, 딥러닝 알고리즘이 데이터만 보고 스스로 중요한 부분을 찾아 분석하도록 함

이전에도 나왔던 딥러닝인데 왜 이제와서 각광받는가?
- 알렉스 넷이라는 알고리즘과 함께 매우 발전했다
- 스마트폰 이후로, 온라인화되어 데이터 수집이 용이해졌다
- 딥러닝을 위해 필요했던 GPU가 좋아졌다 = 높은 컴퓨터 성능
    - 딥러닝은 기본적으로 행렬 연산을 한다! 이러한 연산용 GPU가 생겨나면서 대용량 연산이 가능해졌다 
- 커뮤니티가 발전하여 사람들이 서로 공유하면서 많은 사람들이 유입되기 시작했다 = 개선된 학습방법

## Histprical background of deep learning

딥러닝의 정의
- 머신러닝의 영역, 주로 뉴럴 네트워크 구조를 사용
- 고수준의 추상화를 하는 것
- 많은 layer를 쌓아 non-linear로 변환하는 것

Neural Network
- 인간의 뇌처럼 작동하는 기계를 만들어보다
    - 뉴런에 자극이 올 때, 특정 인계점 즉 threshold를 넘어야지만 그 다음 뉴런으로 자극을 전송하게 되는 그런 특징을 지닌다
- 이를 수식으로 만드는 것
    - 각 input값에 가중치를 linear combination을 거침 -> Activation function을 거쳐 다음 뉴런에게 전달하고자 함

Perceptron
- 인간의 신경(뉴런)을 모방한 머신러닝 알고리즘 
- 각 input값에 가중치를 linear combination을 거칠 경우, 활성함수 (계단 함수)에 의해 0 혹은 1로 결과값이 나온다
    - 활성함수를 시그모이드 함수로 바꾸면 로지스틱 회귀와 유사

- 퍼셉트론은 XOR 문제도 못푼다고 알려짐
    - TF 문제가 주어졌을 때, 해당 문제가 OR인지 AND로 연결되어있는지 알아내는 것

- 걱 NOT, AND, OR, XOR에 따라서 T/F 계산 후, linear 선을 그어 구분해준다
- 그러나 XOR 같은 경우에는 linear이 아닌 곡선으로 표현되어 해당 퍼셉트론으로는 표현이 불가능했다

- 이는 퍼셉트론이 선형분리 (linearly separable) 문제만 풀 수 있기 때문
    - 쉽게 말해 선 하나로 데이터를 분류할 수 있는 문제만 가능
- 그렇다면 해당 퍼셉트론을 여러층을 쌓아보자!

멀티레이어 퍼셉트론
- 말 그대로 층이 여러개인 퍼셉트론 구조
    - 입력층, 은닉층, 출력층으로 이루어짐
- 아래 구조는 은닉층이 1개인 MLP
- 퍼셉트론과 동일하게 활성함수가 레이어마다 존재

예시 : XOR 계산

- 그러나 여기서 가중치 (W) 값들 또한 적절히 찾아야 한다!

Backpropagation
- 체인 룰을 사용한 미분
    - (2x)^3과 같이 괄호 안에 있는 값을.. 치환해서 미분 계산하는 경우를 체인 룰이라고 함

뉴럴넷 학습 방법
- Forward-propagation
    - 입력값이 주어지면 NN(뉴럴넷)을 통과시켜 생성되는 최종 score와 loss 출력
- Backward-propagation
    - 도출된 loss를 이용해 레이어를 거꾸로 순회하면서 NN 학습
    - 미분을 하면서 거꾸로 순회
- 어떻게 진행해야 가중치의 값을 잘 찾을지! 가 뉴럴넷에서 가장 핵심적인 학습 방법이다
    = 이를 경사하강법이라 함
- 네트워크의 학습은 함수 최적화의 방법에 따라 진행
    - Backward-propagation시 gradient descent 알고리즘을 사용하여 학습

History
- 퍼셉트론 (1958) -> XOR 문제 해결 여부 (1969) -> Back-propagation (1970~80s) -> LeNet (1990s) -> AlexNet (2012)
    - LeNet : 숫자를 분석할 수 있는 어플리케이션
