## 확률론

- 딥러닝은 **확률론 기반**의 기계학습 이론에 바탕을 둔다
- 기계학습에서 사용되는 손실함수(loss function)의 작동 원리는 데이터 공간을 통계적으로 해석해서 유도함
    - 회귀분석에서 손실함수로 사용되는 L2노름은 **예측오차의 분산을 가장 최소화하는 방향으로 학습**하도록 유도
    - 분류 문제에서 사용되는 교차엔트로피는 **모델 예측의 불확실성을 최소화하는 방향으로 학습**하도록 유도
    - 여기서 분산 및 불확실성을 최소화하기 위해서는 측정하는 방법을 알아야 한다! 이를 통계학에서 제공하기 때문에 확률론의 기본 개념을 파악해야 함
    - 즉, 회귀든 분류이든 기계학습에서 사용되는 손실함수는 데이터의 분포와 모델 예측의 분포의 차이를 최소화하는 방향으로 학습하게 되는데, 이때 이를 모두 확률론을 기반으로 계산하기 때문에 이를 공부해야 모델에 따라 적절한 손실함수를 사용할 수 있게 된다

확률분포는 데이터의 초상화이다
- 데이터 공간을 XxY라 표기하고 D는 데이터 공간에서 데이터를 추출하는 분포라고 했을 때, 추출된 데이터만 가지고 해당 D를 아는 것은 불가능하기 때문에 기계 학습 모양을 가지고 추론하는 것이다
    - 데이터가 정답 레이블을 항상 가진 지도학습을 상징하기 때문에, 만약 정답 레이블이 없는 경우에는 XxY (Product space)가 아닌 그냥 x공간이라고 표기하게 된다
- 관찰하는 데이터는 확률변수로 이해하며, 확률변수는 데이터는 데이터 공간 상에서 관찰 가능한 즉, 데이터 공간 상의 원소로서 표시하게 된다
    - 확률변수는 함수로 해석하게 되며, 임의로 랜덤하게 데이터 공간안에 관측하게 되는 함수라고 보게 된다
    - 데이터를 추출할 때, 확률변수를 사용하게 되고, 확률변수로 추출한 데이터의 분포를 D라고 한다

$$ (x, y) \sim \mathscr{D} $$

확률변수의 종류
- 확률변수는 확률분포에 따라 이산형과 연속형 확률변수로 구분하게 된다
    - 데이터공간보다는 확률분포 D에 따라서 정해지는 것
    - 실수 공간이라고 해도 확률분포에 따라서 이산형 변수가 나올 수 있다

이산형 확률변수 (확률질량함수)
- 확률변수가 가질 수 있는 경우의 수를 모두 고려하여 확률을 더해서 모델링한다
- 확률변수 X가 x값을 가질 확률

$$ \mathbb{P}(X \in A) = \sum_{x \in A} P(X=x) $$

연속형 확률변수 (확률밀도함수)
- 데이터 공간에 정의된 확률변수의 밀도 위에서의 적분을 통해 모델링한다

$$ \mathbb{P}(X \in A) = \int\limits_A P(X)dx $$

여기서 P(X)는

$$ P(X) = \lim\limits_{h \to 0} \frac{P(x-h)\leq X \leq x+h}{2h}$$

이며, 밀도는 누적확률분포의 변화율을 모델링하는 것
   - 이산형 확률변수와 같이 확률로 해석하면 안된다!
   - 데이터 공간에 정의된 확률보다는 밀도의 개념으로써 적분을 해야 정확히 계산이 가능하다는 점

이산형과 연속형, 항상 두 가지만 모든 확률변수를 표현하는 것은 아니다!
- 대체적으로 표현하는 것이지, 모든 변수가 해당 두 가지로 구분되는 것이 아니다

이렇듯 확률분포에 따라서 모델링하는 방식에 차이가 있으므로, 데이터 접근 방식에 따라 분포의 성질이 달라지게 되고, 분포의 종류에 따라서 확률분포를 모델링하는 방법이 달라지게 된다

- 연속확률분포 모양을 띄고 있지만, 각 빨간색으로 칸을 나누어서 Y=1인 경우, 2인 경우를 세보면 이산형 확률분포로 볼 수 있다
    - 즉, 주어진 데이터의 결합분포를 가지고 원래 확률분포의 D를 모델링할 수 있다
    - 여기서 원래 확률분포 D가 이산형인지 연속형인지에 따라 결합분포를 결정하는 것이 아니라 결합분포를 한정할 때는 원래의 확률분포와 연관되지 않는다!
        - 모델링 방법에 따라 결정되기 때문에 원래 확률분포 D와 설정된 분포는 다를 수 있다는 점을 유의해야 한다
        - 근사하는 방법을 아는 것!

결합분포가 주어진 상황에서 **각 입력 x에 대해 y항에 해당하는 것들을 더해주거나 적분해주게 될 경우, 입력 x에 대한 주변확률분포**를 얻을 수 있다
- P(x)는 입력 X에 대한 주변확률분포로 y에 대한 정보를 주진 않는다
- 즉, 각 X에 따른 분포만 얻을 수 있고, y에 대해 구별할 수는 없다!

$$ P(X) = \sum_{y} P(x,y) \\ P(X) =  \int\limits_y P(x,y)dy$$

- 꼭, x에 대해서만 구하는 것은 아님! 반대로 y에 대해 수행한다면 y에 대한 주변확률분포를 얻을 수 있음!

조건부확률분포 P(X|y)는 데이터 공간에서 입력 x와 출력 y 사이의 관계를 모델링한다

조건부확률분포 구하는 방법
1. y가 주어진 상태에서 x를 구한다
    - 특정 클래스가 주어진 조건에서 데이터의 확률분포를 보여주기 때문에, x와 y의 통계적 관계 모델 혹은 예측 모형을 출력할 때 사용 가능하다
    - 주어진 클래스에 대해 x 분포가 어떻게 형성되어있는지 확인하기 위해서는 주변확률분포보다는 조건부확률분포를 통해 좀 더 명확하게 확인할 수 있게 된다

조건부확률과 기계학습
- 조건부확률 P(y|x)는 입력변수 x에 대해 정답이 y일 확률을 의미한다
    - 연속확률분포의 경우 P(y|x)는 확률이 아닌 밀도로 해석한다는 점 유의!

- 로지스틱 회귀에서 사용했던 선형모델과 소프트맥스 함수의 결합은 데이터에서 추출된 패턴을 기반으로 확률을 해석하는데 사용
    - 분류 문제에서 구해진 패턴에 선형모델을 통해 소프트맥스 함수에 집어넣으면 확률벡터를 구할 수 있었다
    - 데이터로부터 추출된 특징 벡터 Bx과 가중치 행렬 W일 때, 정답이 y일 확률을 계산하는 것은 P(y|x)와 동일하다
    - 즉, 분류 문제에서 softmax(W Phi +b)는 데이터 x로부터 추출된 특징패턴 Phi(x)와 가중치행렬 W를 통해 조건부확률 P(y|x)를 계산한다
- 회귀 문제의 경우, 연속화률변수를 보통 다뤄 확률보단 밀도로 계산해야 함
    - 즉, 조건부기대값 E[y|x]으로 정답을 추론하는 예측모델을 모델링하게 된다

$$ \mathbb{E}_y \sim P_{(y|x)}[y|x] = \int\limits_y yP(y|X)dx $$

    - 만약 이산확률변수인 경우 해당 부분이 적분이 아닌 sum으로 변경!

- 딥러닝은 다중신경망을 사용하여 데이터로부터 특징패턴 phi를 추출하게 된다
    - 특징패턴을 학습하기 위해 어떤 손실함수를 사용할지는 기계학습 문제와 모델에 따라 결정

회귀 문제를 다루는 경우, 손실함수가 L2-norm을 다룬다고 했는데, 조건부기대값은 해당 L2-norm을 최소화하는 f(x)와 일치하기 때문!
- 항상 조건부기대값을 사용하는 것이 아닌 모델에 따라 다른 예측 통계량을 사용할 수 있다!
- 즉, 통계적 목적에 따라 추정량이 달라질 수 있다!

기대값
- 확률분포가 주어지면 데이터를 분석하는데 사용 가능한 여러 종류의 통계적 범함수(statistical functional)를 계산할 수 있다
- 기대값을 **데이터를 대표하는 통계량**이면서 동시에 확률분포를 통해 다른 통계적 범함수를 계산하는데 사용된다
- 모델링하는 확률분포에서 계산해보고 싶은 **다른 통계적 범함수를 사용하는데 이용되는 도구**라고 볼 수 있다
- 기대값을 계산함을 통해 데이터를 해석할 떄 여러 방면에서 활용할 수 있다

$$ \mathbb{E}_X \sim P{(X)}[f(x)] = \int\limits_X f(X)P(X)dx $$
$$ \mathbb{E}_X \sim P{(X)}[f(x)] = \sum_{X \in P(X)} f(X)P(X)$$

해당 기대값을 이용해 분산, 첨도, 공분산 등 여러 통계량 계산 가능

몬테카를로 샘플링 (Monte Carlo)
- 기계학습의 많은 문제들은 확률분포를 명시적으로 모를 때가 대부분이다
- 확률분포를 모를 때 데이터를 이용하여 기대값을 계산하려면 몬테카를로 샘플링 방법을 사용한다
    - 샘플링 방법을 알고 있다면 적분이나 sum 대신 샘플링을 대신하여 기대값을 계산한다
- 이산형, 연속형 상관없다! 대신 분포에서 독립적으로 샘플링을 해줘야만 몬테카를로 샘플링 방법이 작동한다
- 즉, 몬테카를로 샘플링은 독립추출만 보장된다면 대수의 법칙에 의해 수렴성을 보장하며 기대값에 근사하게 된다

$$ \mathbb{E}_X \sim P{(X)}[f(x)] = \frac{1}{N} \sum_{i=1}^{N} f(X^{(i)}),    X^{(i)} \sim^{i.i.d.}P(X)$$

몬테카를로 예제 : 적분 계산하기
$$함수 f(x)=e^{-x^2}의 [-1, 1]상에서 적분값을 어떻게 구할까?$$

$$ \frac{1}{2} \int\limits_{-1}^1 e^{-x^2} dx \approx \frac{1}{N} \sum_{i=1}^{N} f(X^{(i)}),    X^{(i)} \sim U(-1,1)$$

구간 [-1, 1]의 길이는 2이므로 적분값을 2로 나누면 기대값을 계산하는 것과 동일! -> 몬테카를로 샘플링 사용가능
- 균등분포를 사용하는 것과 동일

샘플사이즈가 너무 커지게 될 경우, 오차범위가 커질 수 있어 적절한 개수를 조절해줘야 한다

```python
import numpy as np

def mc_int(fun, low, high, sample_size=100, repeat=10):
    int_len = np.abs(high-low)
    stat = []
    for _ in range(repeat):
        x = np.random.uniform(low=low, high=high, size=sample_size)
        fun_x = fun(x)
        int_val = int_len * np.mean(fun_x)
        stat.append(int_val)
    return np.mean(stat), np.std(stat)

def f_x(x):
    return np.exp(-x**2)

print(mc_int(f_x, low=-1, high=1, sample_size=10000, repeat=100))
```

적분값과 그에 따른 분산을 계산하여 오차범위 또한 구할 수 있다!
- 실제 참 값과 비교했을 때 유사한 것을 확인할 수 있다

## 통계학

- 통계적 모델링은 적절한 가정 위에 확률분포를 추정하는 것이 목표이며, 기계학습과 통계학이 공통적으로 추구하는 목표
- 실제 사용하는 분포도 다양하며, 데이터의 실제 분포를 알 수 없기 떄문에 어떠한 분포를 사용하여 모델링하는지가 중요!
- 유한한 개수의 데이터만 관찰해서 모집단의 분포를 정확하게 알아낸다는 것은 불가능하다. 근사적으로 확률분포를 추정할 수 밖에 없다
    - 예측모형의 목적은 분포를 정확하게 맞추는 것보다는 데이터와 추정 방법의 불확실성을 고려해서 예측의 위험을 최소화하는 방향으로 학습하는 것이기 때문
    - 즉, 추정하는 것만으로도 예측을 할 수 있다!

모수
- 데이터가 특정 확률분포를 따른다고 선험적으로 (a priori) 가정한 후 그 분포를 결정하는 모수를 추정하는 방법을 모수적 (parametric) 방법론이라 한다
    - 만약 정규분포를 가정할 때, 평균과 분산을 모수라고 칭하며 해당 두 모수를 추정하는 방법을 모수적 방법론이라 함
- 특정 롿융분포를 가정하지 않고 데이터에 따라 모델의 구조 및 모수의 개수가 유연하게 바뀌면 비모수 방법론이라 함
    - 기계학습의 많은 방법론은 비모수 방법론에 속함
    - 비모수라고 해서 모수가 존재하지 않는 것이 아닌 모수가 엄청나게 많거나 모수의 개수가 유연하게 바뀔 때 비모수 방법론이라고 함

- 즉, 비모수 & 모수는 어떤 가정을 부여하는지에 따라서 결정된다!

확률분포 가정하기 : 예제
확률분포를 가정하는 방법 -> 우선 히스토그램을 통해 전체적인 모양을 관찰한다

- 주의할 점은 기계적으로 확률분포를 가정하는 것이 아니라, 데이터를 생성하는 원리를 먼저 고려하고 가정을 해야한다
    - 각 분포마다 검정하는 방법들이 있어 모수를 추정한 후에 검정을 할 필요가 있다

데이터로 모수 추정
- 데이터의 확률분포를 가정했다면 모수를 추정해볼 수 있다
- 정규분포의 모수는 평균과 분산


- 표본평균
$$ \bar{X} = \frac{1}{N} \sum_{i=1}^{N} X_{i} $$
$$ \mathbb{E}[\bar{X}] = \mu $$
- 표본분산
$$ S^2 = \frac{1}{N-1} \sum_{i=1}^{N} (X_{i}-\bar{X}) $$
$$ \mathbb{E}[S^2] = \sigma^2 $$

- 여기서 표본분산의 경우 기존 모집단의 분산과 일치하는 불편 추정량을 구하기 위해 N이 아닌 N-1로 나누게 된다
- 이와 같이 정의된 통계량은 모집단의 평균과 분산과 일치하게 된다
- 이를 통해 데이터 확률분포 모수를 추정할 수 있다

표집분포
- 통계량의 확률분포를 표집분포라고 부른다
    - 표본들의 분포가 아닌, 표본평균, 표본분산 등 통계량의 확률분포를 표집분포라고 부르게 된다
    - 표본분포 =/= 표집분포
- 특히 표본평균의 표집분포는 N이 커질수록 정규분포 N(mu, sigma^2/N)을 따르게 된다
    - 이를 중심극한정리(Central Limit Theorem)라 부르며, 모집단의 분포가 정규분포를 따르지 않아도 표집분포는 따른다

최대가능도 추정법
- 표본평균이나 표본분산은 중요하지만, 확률분포에 따라 사용하는 모수가 달라 적절한 통계량이 달라지게 됨
- 이론적으로 가장 가능성이 높은 모수를 추정하는 방법 중 하나는 최대가능도 추정법 (maximum likelihood estimation, MLE)
    - 주어진 확률분포를 어떤식으로 가정하냐와는 관계 없다

$$ \hat{\theta}_{MLE} = 
 \begin{matrix}
  argmax \\
  \theta
 \end{matrix}
L(\theta;x)
=  \begin{matrix}
  argmax \\
  \theta
 \end{matrix}
P(x|\theta)
$$

- 기존에는 모수 theta가 주어졌을 때 x에 대한 함수로 정의하지만, 관점에 차이가 존재
    - 여기서 확률로 이해하는 것이 아니라, 대소비교가 가능한 함수로 이해해야 한다
        - 확률의 특징을 지니고 있는 것이 아니기 때문! (모두 더할 경우 1 등..)
    - 주어진 데이터 x에 대해 모수 theta를 변수로 둔 함수이다
    - 즉, 데이터가 이미 주어진 상태에서 theta를 변화함에 따라 값이 바뀌는 함수

$$ L(\theta;X) = \prod_{i=1}^{n} P(x_i|\theta) \implies logL(\theta;X) = \sum_{i=1}^{n} log P(x_i|\theta) $$

- 데이터 집합 X가 독립적으로 추출되었을 경우 로그가능도를 최적화한다
    - 데이터의 각 행 벡터가 독립적으로 추출 -> 확률밀도 혹은 질량함수들의 곱으로 표현할 수 있다
    - 이때, 로그함수의 성질을 이용해서 덧셈으로 계산하여 최적화할 수 있게 된다

왜 로그가능도를 사용하나요?
- 로그가능도를 최적화하는 모수 theta는 가능도를 최적화하는 MLE가 된다
- 데이터의 숫자가 적으면 상관없지만 만일 데이터의 숫자가 수억 단위가 된다면 컴퓨터로 가능도를 계산하는 것이 불가능함
    - 이를 로그를 사용하여 곱셈을 덧셈으로 변경하게 되면 연산이 가능해진다!
- 경사하강법으로 가능도를 최적화할 때 미분 연산을 사용하게 되는데 미분연산은 데이터의 제곱에 비례하는 연산이 필요하다
    - 로그가능도를 사용하면 연산량이 O(n^2)이 아닌 O(n)으로 복잡도를 선형적으로 변환해주게 된다
- 대개의 손실함수의 경우 경사하강법을 사용하므로 음의 로그가능도 (negative log-likelihood)를 최적화하게 된다
    - 목적식을 최소화하는 목적식으로 변환해주기 위해!

최대가능도 추정법 예제 : 정규분포
- 정규분포를 따르는 확률변수 X로부터 독립적인 표본 [x_1, ...., x_n]을 얻었을 때 최대가능도 추정법을 이용하여 모수를 추정하면?

$$ logL(\theta;X) = \sum_{i=1}^{n} log P(x_i|\theta) $$

- theta = (mu, \sigma^2)에 대해 오른쪽 수식을 미분해서 최적화할 수 있다

- 두 미분이 모두 0이 되는 mu, sigma를 찾으면 가능도를 최대화하게 된다

- MLE를 통해 구한 평균은 표본평균과 동일하지만, 분산은 그렇지 않은 것을 볼 수 있다
    - likelihood를 최소화하지만, MLE는 불편추정량을 보장하지 않는다!'
    - 복잡도를 선형적으로 보장한다는 점
- 즉, 각자 장점에 따라 여러 통계량을 사용할 수 있다

최대가능도 추정법 예제 : 카테고리 분포
- 카테고리 분포 Multinoulli(x; p_1, ...., p_d)를 따르는 확률변수 X로부터 독립적인 표본 [x_1, ...., x_n]을 얻었을 때 최대가능도 추정법을 이용하여 모수를 추정하면
    - 카테고리 분포 : 베르누이 분포 (0과 1 중 하나를 선택하는 분포)를 다차원 d로 확장한 개념
    - 정규분포에서 사용된 모수는 평균과 분산, 카테고리에서의 분포 즉, 확률질량함수 P1, ..., Pd는 각 차원에서 값이 0과 1 중 하나가 될 확률을 의미하기 때문에 P1~Pd를 모두 더해줬을 때, 1이 되어야하는 성질이 있다 

- 주어진 데이터 x_i에서 k번째 차원에 해당하는 값을 추출해주는 형태러 계산하게 된다
    - 만약 x_i, k가 0이 될 경우, P_k = 1이 되며
    - 0이 아닐 경우, P_k만 남기 때문에 (0과 1 중 하나를 택하므로)

- 먼저 데이터 개수만큼 더해주는 덧셈을 진행한 후, 각 차원별로 덧셈을 표현
    - 여기서 데이터 개수만큼 더해주는 것은 각 데이터들에 대해 각 k값이 1인 개수를 카운팅하는 n_k라는 값으로 대체해서 작성 가능
        - n_k는 즉, 주어진 데이터 x_i에 대해서 k값이 1인 데이터 개수를 세는 개념으로 이해

- 여기서 오른쪽 수식을 만족하면서 왼쪽 목적식을 최대화하는 것 = MLE
- 이렇듯 목적식에 제약식이 추가된 상황에서는 미분 = 0이 아닌 라그랑주 승수법을 통해 새로운 제약식을 만들어줄 수 있다

$$ p_k = \frac{n_k}{\sum_{i=1}^{d} n_k} $$
여기서 $$ \sum_{i=1}^{d} n_k = n $$
이므로
- 각 차원의 데이터를 모두 더한 것이므로, 총 데이터의 개수와 동일
$$ p_k = \frac{n_k}{n} $$
- 이는 각 차원의 데이터 개수를 세어 총 데이터 개수와의 비율을 구하는 것으로 최대가능도를 개선하는 모수를 추정하는 것이 가능해진다

딥러닝에서 최대가능도 추정법
- 최대가능도 추정법을 이용해서 기계학습 모델을 학습할 수 있다
- 딥러닝 모델의 가중치를 theta = (W^(1), ..., W^(L))라 표기했을 때 분류문제에서 소프트맥스 벡터는 카테고리분포의 모수 (p_1, ...., p_k)를 모델링한다
    - L번째 레이어, 
- 원핫벡터로 표현한 정답레이블 y=(y_1, ...., y_k)을 관찰데이터로 이용해 확률분포인 소프트맥스 벡터의 로그가능도를 최적화할 수 있다

확률분포의 거리를 구하자
- 기계학습에서 사용되는 손실함수들은 모델이 학습하는 확률분포와 데이터에서 관찰되는 확률분포의 거리를 통해 유도
    - MLE로 추정하게 되는 많은 방법론들이 확률분포의 거리를 최적화하는 것과 밀접하게 관련이 있게 된다
- 데이터 공간에 두 개의 확률분포 P(X), Q(X)가 있을 경우 두 확률분포 사이의 거리를 계산할 때 다음과 같은 함수들 사용
    - 총변동거리, 쿨백-라이블러 발산, 바슈타인 거리

쿨백-라이블러 발산(KL Divergence)

- 즉, 확률분포 사이의 거리를 최소화시키는 개념과 가능도함수를 최대화한다는 개념이 밀접하다
    - 결국 기계학습의 원리가 확률분포 사이의 거리를 최소화하는 것과 동일함을 알 수 있다
